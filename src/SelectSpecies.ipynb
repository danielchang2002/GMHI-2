```python
import pandas as pd
df = pd.read_csv("4347_final_relative_abundances.txt", sep='\t')
df
```

```python
df.columns = ['SampleID'] + list(df.columns[1:])
df
```

```python
transposed = df.T
transposed.head()
```

```python
new_header = transposed.iloc[0].copy()
final = transposed[1:] #take the data less the header row
final.columns = new_header #set the header row as the df header
final.head()
```

```python
theta_f = 1.366
theta_d = 0.104
```

```python
meta = pd.read_excel("41467_2020_18476_MOESM3_ESM.xlsx")
isHealthy = list(meta.iloc[2:-1, 7] == 'Healthy')
```

```python
sum(isHealthy)
```

```python
healthy = final.loc[isHealthy]
healthy
```

```python
notHealthyArray = [not h for h in isHealthy]
notHealthy = final.loc[notHealthyArray]
notHealthy
```

```python
thresh = 0.00001
p_h = (healthy > thresh).sum(axis=0) / len(healthy)
proportion_healthy = pd.DataFrame(p_h)
proportion_healthy.head()
```

```python
p_n = (notHealthy > thresh).sum(axis=0) / len(notHealthy)
proportion_not = pd.DataFrame(p_n)
proportion_not.head()
```

```python
fold_change_healthy = proportion_healthy / proportion_not
fold_change_healthy.head()
```

```python
difference_healthy = proportion_healthy - proportion_not
difference_healthy.head()
```

```python
vals1 = pd.DataFrame({"fold" : fold_change_healthy.iloc[:, 0], 
                     "diff" : difference_healthy.iloc[:, 0]})
vals1
```

```python
health_abundant = vals1[vals1['fold'] >= theta_f]
health_abundant = health_abundant[health_abundant['diff'] >= theta_d]
health_abundant
```

```python
fold_change_not = proportion_not / proportion_healthy
fold_change_not.head()
```

```python
difference_not = proportion_not - proportion_healthy
difference_not.head()
```

```python
vals2 = pd.DataFrame({"fold" : fold_change_not.iloc[:, 0], 
                     "diff" : difference_not.iloc[:, 0]})
vals2
```

```python
health_scarce = vals2[vals2['fold'] >= theta_f]
health_scarce = health_scarce[health_scarce['diff'] >= theta_d]
print(len(health_scarce))
health_scarce.head()
```
