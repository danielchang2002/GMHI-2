## Setup

```python
import pandas as pd
import config
```

```python
species = pd.read_csv(config.TRAIN_DIR + "taxonomy.csv", index_col = 0)
species.head()
```

```python
isHealthy = pd.read_csv(config.TRAIN_DIR + "isHealthy.csv", index_col=0)
isHealthy
```

```python
sum(isHealthy.iloc[0, :])
```

```python
# Best parameters found in the paper
theta_f = 1.4
theta_d = 0.1
```

```python
full = pd.concat([species.T, isHealthy.T], axis=1)
full
```

## Get Healthy and Unhealthy dataframes

```python
healthySamples = full[full['isHealthy']].iloc[:, :-1]
healthySamples
```

```python
notSamples = full[~full['isHealthy']].iloc[:, :-1]
notSamples
```

```python
sum(list(healthySamples.iloc[0, :]))
```

## Get proportions for healthy and unhealthy tables

```python
thresh = 0.00001 # 1 * 10e-5
```

```python
p_h = (healthySamples > thresh).sum() / len(healthySamples)
proportionHealthy = pd.DataFrame({"proportion" : p_h})
proportionHealthy
```

```python
p_n = (notSamples > thresh).sum() / len(notSamples)
proportionNot = pd.DataFrame({"proportion" : p_n})
proportionNot
```

## Select health abundant species, $M_H$

```python
fold_change_healthy = proportionHealthy / proportionNot
fold_change_healthy.head()
```

```python
difference_healthy = proportionHealthy - proportionNot
difference_healthy.head()
```

```python
healthyTable = pd.DataFrame({"fold" : fold_change_healthy.iloc[:, 0], 
                     "diff" : difference_healthy.iloc[:, 0]})
healthyTable
```

```python
health_abundant = healthyTable[healthyTable['fold'] >= theta_f]
health_abundant = health_abundant[health_abundant['diff'] >= theta_d]
health_abundant
```

```python
health_abundant_list = list(health_abundant.index)
health_abundant_list
```

```python
# Write to file
with open(config.OUTPUT_DIR + "MH_species.txt", 'w') as f:
    for species in health_abundant_list:
        f.write(species + '\n')
```

## Select health scarce species, $M_N$

```python
fold_change_not = proportionNot / proportionHealthy
fold_change_not.head()
```

```python
difference_not = proportionNot - proportionHealthy
difference_not.head()
```

```python
notTable = pd.DataFrame({"fold" : fold_change_not.iloc[:, 0], 
                     "diff" : difference_not.iloc[:, 0]})
notTable
```

```python
health_scarce = notTable[notTable['fold'] >= theta_f]
health_scarce = health_scarce[health_scarce['diff'] >= theta_d]
print(len(health_scarce))
health_scarce.head()
```

```python
health_scarce_list = list(health_scarce.index)
# Write to file
with open("../output/MN_species.txt", 'w') as f:
    for species in health_scarce_list:
        f.write(species + '\n')
```
