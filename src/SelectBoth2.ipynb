## Setup

```python
data_dir = "../data/"
```

```python
import pandas as pd
pathways = pd.read_csv(data_dir + "pathways.csv", index_col = 0)
pathways
```

```python
species = pd.read_csv(data_dir + "taxonomy.csv", index_col = 0)
species.head()
```

```python
isHealthy = pd.read_csv(data_dir + "isHealthy.csv", index_col=0)
isHealthy
```

```python
sum(isHealthy.iloc[0, :])
```

```python
full = pd.concat([pathways.T, species.T, isHealthy.T], axis=1)
full
```

```python
healthy = full[full['isHealthy']].iloc[:, :-1]
healthy
```

```python
notHealthy = full[~full['isHealthy']].iloc[:, :-1]
notHealthy
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(
degree=2, interaction_only=True, include_bias=False
)
```

```python
pf.fit(healthy)
poly_feats = pf.transform(healthy)
healthy_transformed = pd.DataFrame(
    poly_feats,
    columns=pf.get_feature_names(healthy.columns),
    index=healthy.index
)
healthy_transformed
```

```python
pf.fit(notHealthy)
poly_feats = pf.transform(notHealthy)
notHealthy_transformed = pd.DataFrame(
    poly_feats,
    columns=pf.get_feature_names(notHealthy.columns),
    index=notHealthy.index
)
notHealthy_transformed
```

## Get Averages

```python
healthy_transformed.shape[0]
```

```python
healthy_transformed.sum()
```

```python
0.077275 / 2636
```

```python
averageH = healthy_transformed.sum() / healthy_transformed.shape[0]
averageHealthy = pd.DataFrame({"average" : averageH})
averageHealthy
```

```python
averageN = notHealthy_transformed.sum() / notHealthy_transformed.shape[0]
averageNot = pd.DataFrame({"average" : averageN})
averageNot
```

## Select health abundant pathways, $B_H$

```python
fold_change_healthy = averageHealthy / averageNot
fold_change_healthy
```

```python
difference_healthy = averageHealthy - averageNot
difference_healthy.head()
```

```python
healthyTable = pd.DataFrame({"fold" : fold_change_healthy.iloc[:, 0], 
                     "diff" : difference_healthy.iloc[:, 0]})
healthyTable
```

```python
theta_f = 1.2
theta_d = 0.003
```

```python
health_abundant = healthyTable[healthyTable['fold'] >= theta_f]
health_abundant = health_abundant[health_abundant['diff'] >= theta_d]
health_abundant
```

```python
health_abundant_list = list(health_abundant.index)
health_abundant_list
```

```python
# Write to file
with open("../output/BothHealthy.txt", 'w') as f:
    for pathway in health_abundant_list:
        f.write(pathway + '\n')
```

## Select health scarce pathways, $B_N$

```python
fold_change_not = averageNot / averageHealthy
fold_change_not.head()
```

```python
difference_not = averageNot - averageHealthy
difference_not.head()
```

```python
notTable = pd.DataFrame({"fold" : fold_change_not.iloc[:, 0], 
                     "diff" : difference_not.iloc[:, 0]})
notTable
```

```python
health_scarce = notTable[notTable['fold'] >= theta_f]
health_scarce = health_scarce[health_scarce['diff'] >= theta_d]
print(len(health_scarce))
health_scarce
```

```python
health_scarce_list = list(health_scarce.index)
health_scarce_list[:10]
```

```python
# Write to file
with open("../output/BothUnhealthy.txt", 'w') as f:
    for pathway in health_scarce_list:
        f.write(pathway + '\n')
```
