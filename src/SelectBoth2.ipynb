## Setup

```python
data_dir = "../data/"
```

```python
import pandas as pd
pathways = pd.read_csv(data_dir + "pathways.csv", index_col = 0)
pathways
```

```python
species = pd.read_csv(data_dir + "taxonomy.csv", index_col = 0)
species.head()
```

```python
isHealthy = pd.read_csv(data_dir + "isHealthy.csv", index_col=0)
isHealthy
```

```python
sum(isHealthy.iloc[0, :])
```

```python
full = pd.concat([pathways.T, species.T], axis=1)
full = pd.concat([full, isHealthy.T], axis=1)
full
```

## Get Healthy and Unhealthy dataframes

```python
healthySamples = full[full['isHealthy']].iloc[:, :-1]
healthySamples
```

```python
notSamples = full[~full['isHealthy']].iloc[:, :-1]
notSamples
```

```python
sum(list(healthySamples.iloc[0, :]))
```

## Get averages for each feature

```python
average_h = healthySamples.sum() / len(healthySamples)
averageHealthy = pd.DataFrame({"average" : average_h})
averageHealthy
```

```python
average_n = notSamples.sum() / len(notSamples)
averageNot = pd.DataFrame({"average" : average_n})
averageNot
```

## Select health abundant pathways, $B_H$

```python
fold_change_healthy = averageHealthy / averageNot
fold_change_healthy.head()
```

```python
difference_healthy = averageHealthy - averageNot
difference_healthy.head()
```

```python
healthyTable = pd.DataFrame({"fold" : fold_change_healthy.iloc[:, 0], 
                     "diff" : difference_healthy.iloc[:, 0]})
healthyTable
```

```python
theta_f = 1.4
theta_d = 0.003
```

```python
health_abundant = healthyTable[healthyTable['fold'] >= theta_f]
health_abundant = health_abundant[health_abundant['diff'] >= theta_d]
health_abundant
```

```python
health_abundant_list = list(health_abundant.index)
health_abundant_list
```

```python
# Write to file
with open("../output/BothHealthy.txt", 'w') as f:
    for pathway in health_abundant_list:
        f.write(pathway + '\n')
```

## Select health scarce pathways, $B_N$

```python
fold_change_not = averageNot / averageHealthy
fold_change_not.head()
```

```python
difference_not = averageNot - averageHealthy
difference_not.head()
```

```python
notTable = pd.DataFrame({"fold" : fold_change_not.iloc[:, 0], 
                     "diff" : difference_not.iloc[:, 0]})
notTable
```

```python
health_scarce = notTable[notTable['fold'] >= theta_f]
health_scarce = health_scarce[health_scarce['diff'] >= theta_d]
print(len(health_scarce))
health_scarce
```

```python
health_scarce_list = list(health_scarce.index)
health_scarce_list[:10]
```

```python
# Write to file
with open("../output/BothUnhealthy.txt", 'w') as f:
    for pathway in health_scarce_list:
        f.write(pathway + '\n')
```
