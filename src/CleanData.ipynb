This notebook converts the files in the config.RAW_DIR into csv files in their corresponding directories. The shapes of the data matrices are n x m, where n = # of features and m = # of samples
The shapes of the label data frames are 1 x m.

## Setup

```python
import pandas as pd
import config
```

## Clean Taxonomy Data

```python
df = pd.read_excel(config.RAW_DIR + "Final_taxonomy_4347.xlsx", engine="openpyxl")
```

```python
# set column names to sample ID row
df2 = df.copy()
df2.columns = list(df2.iloc[6, :])
df2
```

```python
taxonomy = df2.iloc[32:, :]
taxonomy = taxonomy.rename(columns={'Sample Accession or Sample ID': 'Species'})
taxonomy.reset_index(inplace=True, drop=True)
taxonomy = taxonomy.set_index("Species")
taxonomy
```

#### For each sample, the sum of all the values = 100 because the values are relative abundances, as percentages. Scale values by 100 so sum = 1

```python
sum(taxonomy.iloc[:, 3])
```

```python
taxonomy_scaled = taxonomy / 100
taxonomy_scaled.head()
```

```python
sum(taxonomy_scaled.iloc[:, 3])
```

#### Lexicographically sort samples

```python
taxonomy_scaled = taxonomy_scaled.reindex(sorted(taxonomy_scaled.columns), axis=1)
taxonomy_scaled.head()
```

```python
taxonomy_scaled.to_csv(config.TRAIN_DIR + "taxonomy.csv")
```

## Get Training Labels

```python
phenotype = df2.iloc[[14], :]
phenotype.reset_index(inplace=True, drop=True)
phenotype2 = phenotype.set_index("Sample Accession or Sample ID")
phenotype2.index.name = None
isHealthy = phenotype2 == "Healthy"
isHealthy.index = ["isHealthy"]
isHealthy = isHealthy.reindex(sorted(isHealthy.columns), axis=1)
isHealthy
```

```python
isHealthy.to_csv(config.TRAIN_DIR + "isHealthy.csv")
```

```python
sorted_phenotype2 = phenotype2.reindex(sorted(phenotype2.columns), axis=1)
sorted_phenotype2
```

```python
sorted_phenotype2.to_csv(config.TRAIN_DIR + "phenotype.csv")
```

## Clean Pathway Data

```python
df = pd.read_excel(config.RAW_DIR + "Final_MetaCyc_pathways_4347.xlsx", engine="openpyxl")
```

```python
df2 = df.copy()
df2.head()
```

```python
indexed = df2.set_index("Sample Accession or Sample ID")
indexed.head()
```

```python
pathways_cropped = indexed.iloc[:, 30:]
pathways_cropped.index.name = None
pathways_cropped.head()
```

```python
transposed = pathways_cropped.T
transposed.index.name = "Pathway"
transposed
```

```python
sum(transposed.iloc[:, 3])
```

#### Lexicographically Sort Samples

```python
transposed = transposed.reindex(sorted(transposed.columns), axis=1)
transposed.head()
```

```python
transposed.to_csv(config.TRAIN_DIR + "pathways.csv")
```
