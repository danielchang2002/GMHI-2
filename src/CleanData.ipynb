## Setup

```python
data_dir = "../data/"
```

## Clean Taxonomy Data

```python
import pandas as pd
df = pd.read_excel(data_dir + "Final_taxonomy_4347.xlsx")
```

```python
# set column names to sample ID row
df2 = df.copy()
df2.columns = list(df2.iloc[6, :])
df2
```

```python
taxonomy = df2.iloc[32:, :]
taxonomy = taxonomy.rename(columns={'Sample Accession or Sample ID': 'Species'})
taxonomy.reset_index(inplace=True, drop=True)
taxonomy = taxonomy.set_index("Species")
taxonomy
```

#### For each sample, the sum of all the values = 100 because the values are relative abundances, as percentages. Scale values by 100 so sum = 1

```python
sum(taxonomy.iloc[:, 3])
```

```python
taxonomy_scaled = taxonomy / 100
taxonomy_scaled.head()
```

```python
sum(taxonomy_scaled.iloc[:, 3])
```

#### Lexicographically sort samples

```python
taxonomy_scaled = taxonomy_scaled.reindex(sorted(taxonomy_scaled.columns), axis=1)
taxonomy_scaled.head()
```

```python
taxonomy_scaled.to_csv(data_dir + "taxonomy.csv")
```

## Clean Pathway Data

```python
df = pd.read_excel(data_dir + "Final_MetaCyc_pathways_4347.xlsx")
```

```python
df2 = df.copy()
df2.head()
```

```python
indexed = df2.set_index("Sample Accession or Sample ID")
indexed.head()
```

```python
pathways_cropped = indexed.iloc[:, 31:]
pathways_cropped.index.name = None
pathways_cropped.head()
```

```python
transposed = pathways_cropped.T
transposed.index.name = "Pathway"
transposed
```

```python
sum(transposed.iloc[:, 3])
```

#### Lexicographically Sort Samples

```python
transposed = transposed.reindex(sorted(transposed.columns), axis=1)
transposed.head()
```

```python
transposed.to_csv(data_dir + "pathways.csv")
```
