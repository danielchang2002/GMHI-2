## Setup

```python
data_dir = "../data/"
output_dir = "../output/"
```

```python
import pandas as pd
df = pd.read_csv(data_dir + "taxonomy.csv", index_col = 0)
df.head()
```

```python
species = df.copy()
species[species < 0.00001] = 0
species.head()
```

```python
import pandas as pd
df = pd.read_csv(data_dir + "pathways.csv", index_col = 0)
pathways = df.copy()
pathways[pathways < 0.00001] = 0
pathways.head()
```

```python
with open(output_dir + "MH_species.txt") as f:
    MH_species = f.read().splitlines()
with open(output_dir + "MN_species.txt") as f:
    MN_species = f.read().splitlines()
MH_species
```

```python
with open(output_dir + "BH_pathways.txt") as f:
    BH_pathways = f.read().splitlines()
with open(output_dir + "BN_pathways.txt") as f:
    BN_pathways = f.read().splitlines()
BH_pathways
```

```python
MH_species_metagenome = species[species.index.isin(MH_species)]
MN_species_metagenome = species[species.index.isin(MN_species)]
display(MH_species_metagenome)
display(MN_species_metagenome.head())
```

```python
BH_pathways_metagenome = pathways[pathways.index.isin(BH_pathways)]
BN_pathways_metagenome = pathways[pathways.index.isin(BN_pathways)]
display(BH_pathways_metagenome.head())
display(BN_pathways_metagenome.head())
```

```python
import numpy as np

def shannon(df):
    shannoned = np.log(df) * df * -1
    filled = shannoned.fillna(0)
    sums = list(filled.sum(axis=0, ))
    columns = list(df.columns)
    frame = pd.DataFrame(pd.Series(data=sums, index=columns)).T
    frame.index = ['shannon']
    return frame
```

```python
MH_shannon = shannon(MH_species_metagenome)
MN_shannon = shannon(MN_species_metagenome)
MH_shannon.iloc[:, :10]
```

```python
BH_shannon = shannon(BH_pathways_metagenome)
BN_shannon = shannon(BN_pathways_metagenome)
BH_shannon.iloc[:, :10]
```

```python
def richness(df):
    frame = pd.DataFrame((df > 0).sum()).T
    frame.index = ['Richness']
    return frame
```

```python
R_MH = richness(MH_species_metagenome)
R_MN = richness(MN_species_metagenome)
R_MH.iloc[:, :10]
```

```python
R_BH = richness(BH_pathways_metagenome)
R_BN = richness(BN_pathways_metagenome)
R_BH.iloc[:, :10]
```

```python
MH_prime = 7
MN_prime = 31
```

```python
BH_prime = 8
BN_prime = 42
```

```python
psi_MH = pd.DataFrame((R_MH / MH_prime).values * MH_shannon.values, 
                      columns=MH_shannon.columns, index=["psi_MH"])
psi_MN = pd.DataFrame((R_MN / MN_prime).values * MN_shannon.values, 
                      columns=MN_shannon.columns, index=["psi_MN"])
psi_MN
```

```python
psi_BH = pd.DataFrame((R_BH / BH_prime).values * BH_shannon.values, 
                      columns=BH_shannon.columns, index=["psi_BH"])
psi_BN = pd.DataFrame((R_BN / BN_prime).values * BN_shannon.values, 
                      columns=BN_shannon.columns, index=["psi_BN"])
psi_BN
```

```python
def GMHI(psi_MH, psi_MN, psi_BH, psi_BN, lambda1, lambda2):
    num = 0.0001 + psi_MH.values + psi_BH.values * lambda1
    den = 0.0001 + psi_MN.values + psi_BN.values * lambda2
    GMHI = pd.DataFrame(np.log10(num / den),
                        columns=psi_MH.columns, index=["GMHI"]).T
    return GMHI
```

```python
test = GMHI(psi_MH, psi_MN, psi_BH, psi_BN, 0, 0)
test
```

```python
isHealthy = pd.read_csv(data_dir + "isHealthy.csv", index_col=0)
isHealthy.T.head()
```

```python
def balanced_accuracy(GMHI, y):
    full = pd.concat([GMHI, y], axis=1)
    healthies = full[full['isHealthy']].iloc[:, [0]]
    predictedHealthy = int((healthies > 0).sum())
    unhealthies = full[~full['isHealthy']].iloc[:, [0]]
    predictedNot = int((unhealthies < 0).sum())
    P_H = predictedHealthy / len(healthies)
    P_N = predictedNot / len(unhealthies)
    chi = 0.5 * (P_H + P_N)
    return chi
```

```python
balanced_accuracy(test, isHealthy.T)
```

## Lambda = 0 -> Balanced Accuracy of 0.697 hurray

```python
info = []
for i in range(100):
    lambda1 = i * 0.1
    for j in range(100):
        lambda2 = j * 0.1
        test = GMHI(psi_MH, psi_MN, psi_BH, psi_BN, lambda1, lambda2)
        acc = balanced_accuracy(test, isHealthy.T)
        info.append({"lambda1" : lambda1, "lambda2" : lambda2, "acc" : acc})
```
