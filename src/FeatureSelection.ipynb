## Use Lasso Regression to select features

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold

import numpy as np

strat_kfold = StratifiedKFold(10, shuffle=True, random_state=42)

for C in [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30, 100]:
    print(C)
    logreg = LogisticRegression(C=C, penalty='l1',
    fit_intercept=True, max_iter=700, random_state=42,
                        solver='saga', n_jobs=-1)
    scores = cross_val_score(logreg, X, np.ravel(y), cv = strat_kfold, 
                             scoring = "balanced_accuracy")
    print(scores, np.mean(scores))
```

```python
logreg = LogisticRegression(C=1, penalty='l1',
    fit_intercept=True, max_iter=1000, random_state=42,
                        solver='saga', n_jobs=-1)
logreg.fit(X, np.ravel(y))
logreg.coef_[0]
```

```python
coef = pd.DataFrame(X.columns)
coef['coef'] = logreg.coef_[0]
coef
```

```python
nonzero = coef[coef['coef'].abs() > 0.001]
nonzero
```

```python
features = list(nonzero['Species'])
features
```

```python
with open(config.OUTPUT_DIR + "features.txt", 'w') as f:
    for feat in features:
        f.write(feat + "\n")
```
