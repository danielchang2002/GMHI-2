## Demo GHMI2!!!

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from GMHI2 import GMHI2

# Instantiate
model = GMHI2()

# Use 10 fold cv, shuffle data
kfold = KFold(10, shuffle=True, random_state=42)
bal_accs = cross_val_score(model, X_train, y_train, cv=kfold, scoring = "balanced_accuracy")
display(bal_accs, np.mean(bal_accs))
```

#### 10 fold cross validation on training set achieves 70.1% balanced accuracy

```python
from sklearn.metrics import balanced_accuracy_score
from sklearn.metrics import accuracy_score

model.fit(X_train, y_train)
y_val_hat = model.predict(X_val)
acc = accuracy_score(y_val_hat, y_val)
bal_acc = balanced_accuracy_score(y_val_hat, y_val)
acc, bal_acc
```

#### 64.6% balanced accuracy on validation set
